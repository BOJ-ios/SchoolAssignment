# Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

## Dataset

```python
# Load dataset file
data = pd.read_csv('C:/Users/LoewllZoe/Documents/GitHub/SchoolAssignment/231005_Linear_Regression_실습_기본소스코드/kc_house_data.csv', usecols=['price', 'sqft_living'])
data
```

```python
# Numpy 배열로 전환
data_np = np.array(data)

x = np.expand_dims(data_np[:, 1],1)          # sqft_living
y = data_np[:, 0]          # price

# Dataset 시각화
fig = plt.figure(figsize=(8,6))
plt.scatter(x, y, color='b', marker='o', s=30)
plt.show()
```

```python
# Dataset 정규화
x_mean = np.mean(x)
y_mean = np.mean(y)
x_std = np.std(x)
y_std = np.std(y)

x = (x-x_mean)/x_std
y = (y-y_mean)/y_std

# Train dataset / Test dataset 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)
```

## Model


```python
class LinearRegression():
    def __init__(self, learning_rate=0.001, n_iters=1000):
        # init
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weight = None
        self.bias = None

    def fit(self, x, y):
        #Init weight
        self.weight = np.zeros(1)
        self.bias = 0
        
        n_samples, _ = x.shape
        
        #Learning
        for i in range(self.n_iters):
            y_pred = np.dot(x, self.weight)+self.bias
            
            dw = (1/n_samples) * np.sum(np.dot((y_pred-y),x))
            db = (1/n_samples) * np.sum(y_pred-y)
        
            # Update weights
            self.weight = self.weight - self.lr *dw
            self.bias = self.bias - self.lr * db

    def predict(self, x):
        # Prediction
        y_pred = np.dot(x, self.weight)+self.bias
        return y_pred
```

## Prediction

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# loss
loss = np.sum((y_test - y_pred)**2)/len(y_test)
print(loss)
```

```python
# Visualization
fig = plt.figure(figsize=(8,6))
plt.scatter(x_test, y_test, color='b', marker='o', s=30)
plt.plot(x_test, y_pred, 'r')
plt.show()
```

```python
print(model.weight)
print(model.bias)
```

```python
# sampling data
x_sampling = x_test[:10]
y_sampling = y_test[:10]

y_pred = model.predict(x_sampling)

# inverse normalization
x_sampling = (x_sampling * x_std) + x_mean
y_sampling = (y_sampling * y_std) + y_mean
y_pred = (y_pred * y_std)+y_mean

fig = plt.figure(figsize=(8,6))
plt.scatter(x_sampling, y_sampling, color='b', marker='o', s=30)
plt.scatter(x_sampling, y_pred, color='r', marker='o', s=30)
plt.show()
```

```python
# class LinearRegression():
#     def __init__(self, learning_rate=0.001, n_iters=1000):
#         # init
#         self.lr = learning_rate
#         self.n_iters = n_iters
#         self.weight = None
#         self.bias = None

#     def fit(self, x, y):
#         #Init weight
#         self.weight = np.zeros(2)
#         self.bias = 0
        
#         n_samples, _ = x.shape
        
#         #Learning
#         for i in range(self.n_iters):
#             y_pred = np.dot(x, self.weight)+self.bias
            
#             dw = (1/n_samples) * np.sum(np.dot((y_pred-y),x))
#             db = (1/n_samples) * np.sum(y_pred-y)
        
#             # Update weights
#             self.weight = self.weight - self.lr *dw
#             self.bias = self.bias - self.lr * db

#     def predict(self, x):
#         # Prediction
#         y_pred = np.dot(x, self.weight)+self.bias
#         return y_pred
class LinearRegression():
    def __init__(self):
        self.theta = None
    
    def fit(self,x,y):
        n_samples, n_features = x.shape
        x = np.concatenate((x, np.expand_dims(np.ones(x.shape[0]),1)),1)
        
        # Ordinary Least Square method
        self.theta = np.dot(np.linalg.inv(np.dot(x.T, x)),np.dot(x.T,y))
        
    def predict(self, x):
        # Prediction
        x = np.concatenate((x, np.expand_dims(np.ones(x.shape[0]),1)),1)
        y = np.dot(x, self.theta)
        return y
```

## Prediction

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# loss
loss = np.sum((y_test - y_pred)**2)/len(y_test)
print(loss)
```
